## Step 9: Results Analysis and Comparison

In this final step, the performance of all implemented models is compared to evaluate their effectiveness in pneumonia detection.

The goal is to analyze how different approaches—from simple models to advanced transfer learning—impact accuracy and generalization. This comparison highlights the importance of model architecture, training strategy, and data handling in solving complex medical imaging problems.

The insights derived from this analysis provide a comprehensive understanding of model behavior and guide the selection of the most effective approach.


```python
# Model comparison

import pandas as pd

results = {
    "Model": [
        "MLP (Baseline)",
        "CNN (Basic)",
        "CNN (Improved)",
        "Transfer Learning (Frozen)",
        "Transfer Learning (Fine-Tuned)"
    ],
    "Accuracy": [
        0.75,
        0.73,
        0.72,
        0.75,
        0.82
    ]
}

df_results = pd.DataFrame(results)
df_results
```

```python
import matplotlib.pyplot as plt

plt.bar(df_results["Model"], df_results["Accuracy"])
plt.xticks(rotation=30)
plt.ylabel("Accuracy")
plt.title("Model Comparison")
plt.show()
```

## Key Observations

* The baseline MLP model achieved moderate performance despite lacking spatial awareness, demonstrating the presence of learnable patterns in the data.
* Initial CNN models did not outperform the baseline due to insufficient architecture depth and lack of proper training strategies.
* Increasing model complexity without addressing overfitting resulted in degraded performance.
* Transfer learning without fine-tuning provided limited improvement, as pretrained features were not fully adapted to the medical imaging domain.
* The fine-tuned transfer learning model achieved the highest accuracy, highlighting the importance of controlled training, regularization, and data augmentation.

These observations emphasize that model performance depends not only on architecture but also on proper training strategies and data handling.

Starting from a simple baseline model, multiple approaches were explored, including convolutional neural networks and transfer learning. Through systematic experimentation, it was observed that increasing model complexity alone does not guarantee better performance.

## Insights

* Spatial feature extraction is critical for image-based classification tasks, making CNNs more suitable than traditional MLP models.
* Overfitting is a major challenge in medical imaging due to limited data, requiring careful regularization and augmentation techniques.
* Transfer learning significantly improves performance when combined with fine-tuning and domain-specific adaptation.
* Model performance is influenced more by training strategy and data quality than by model complexity alone.
* Controlled experimentation and iterative refinement are essential for building robust machine learning systems.

These insights provide valuable guidance for future work in medical image analysis and deep learning applications.
